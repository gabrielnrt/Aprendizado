```python
import matplotlib.pyplot as plt                                                 
import numpy as np

## Create functions and set domain length
x = np.arange(0.0, 2.0, 0.01)
y1 = x**2 - 0.5
y2 = 2*x - 1

## Plot functions and a point where they intersect
plt.plot(x, y1, label = 'y = x² -0,5', color = 'cyan')
plt.plot(x, y2, label = '2x -1', color = 'black')


## Config the graph
plt.title('Gráfico de duas funções')
plt.xlabel('x')
plt.ylabel('y')


plt.grid(True)
plt.legend(loc = 'lower right')

## Show the graph
plt.show()
```
