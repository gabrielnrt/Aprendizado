# **`histplot()`**

## **Introdução**

Neste notebook abordaremos sobre a função `histplot()` do seaborn, que é importada com o comando

```python
from seaborn import histplot
```

A documentação contendo todos os parâmetros se encontra [aqui](https://seaborn.pydata.org/generated/seaborn.histplot.html#seaborn.histplot).

Essa função realiza um histograma de uma amostra fornecida uma tabela contendo os dados. 

Para ilustrar, usaremos o conjunto de dados da documentação, que no caso é dos penguins, obtidos através de 

```python
from seaborn import load_dataset

df = load_dataset('penguins')

df.dtypes
```

## **Parâmetros Fundamentais**

Os parâmetros fundamentais dessa função são

* `data`: DataFrame ou um array 2D que será referenciado no parâmetro `x`.

* `x`: Nome de uma coluna (string) da tabela em que realizaremos o histograma.

* `bins`: Número de divisões que o histograma terá. Por padrão, está configurado como `'auto'`, onde a própria função determinda automaticamente um número de divisões.

* `kde`: Este parâmetro na verdade não é fundamental. Se colocamos como `True`, então a função fará uma estimativa de densidade kernel (KDE), e se estiver como `False`, então ele não irá realizar essa estmativa. Por padrão, está como `False`.

Realizemos um histograma com respeito à coluna `flipper_length_mm` deixando o número de divisões automático.

```python
histplot(data = df, x = 'flipper_length_mm')
```

Configurando para que o histograma tenha 20 divisões e um gráfico de KDE acompahado, temos

```python
histplot(data = df, x = 'flipper_length_mm', bins = 20, kde = True)
```
