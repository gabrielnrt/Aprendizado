# **`regplot`**

## **Introdução**

Neste notebook vou abordar sobre como realizar um gráfico que represente uma regressão linear de uma amostra. Deste modo, fica implícito que estamos trabalhando com atributos numéricos, tanto no eixo x quanto y. A documentação completa se encontra [neste link](https://seaborn.pydata.org/generated/seaborn.regplot.html).

Para ilustrar a função, usaremos o conjunto de dados de gorjetas, que é obtido através de

```python
from seaborn import load_dataset

df = load_dataset('tips')

df.sample(5)
```

Como podemos observar, as únicas colunas com variáveis numéricas são `total_bill`, `tip` e `size`. Seguindo o exemplo da documentação, usarei as duas primeiras colunas para ilustrar a regressão linear.

Para usarmos a função, devemos chamá-la através de 

```python
from seaborn import regplot
```

## **Parâmetros Fundamentais**

Os parâmetros fundamentais do `regplot()` são:

* `data`: tabela do tipo DataFrame ou array contendo os dados a serem "graficados".

* `x`: Nome da coluna (string) ou `pandas.Series` referente aos dados do eixo x. Note que se passo uma string nesse parâmetro, devo obrigatoriamente passar o parâmetro `data`.

* `y`: Nome da coluna (string) ou `pandas.Series` referente aos dados do eixo y. Note que se passo uma string nesse parâmetro, devo obrigatoriamente passar o parâmetro `data`.

A função retorna o objeto `Axes` do matplotlib, com uma figura contendo os pontos da amostra e o gráfico da regressão linear.

```python
regplot(data = df, x = 'total_bill', y = 'tip')
```
