# **`scatterplot`**

## **Introdução**

Este notebook versará sobre como fazer gráficos de dispersão usando a função `scatterplot` do seaborn. A documentação contendo todos os parâmetros se encontra [neste link](https://seaborn.pydata.org/generated/seaborn.scatterplot.html). 

## **Dados Iniciais**

```python
from pandas import DataFrame

dicionario = {'ValoresX': [0,0,3,-1,5],
              'y1': [-1,-0.5,1,1,2],
              'y2': [0,0,0,0,0]}

df = DataFrame(dicionario)
```

## **Parâmetros Fundamentais**

Para usar a função, basta invocá-la digitando

```python
from seaborn import scatterplot
```

Dentre os parâmetros que essa função possui, destacamos primeiramente:

* `data`: Nome da tabela a ser usada, que pode ser um DataFrame ou array.

* `x`: Nome da coluna da tabela que será usada como eixo x do gráfico.

* `y`: Nome da coluna da tabela que será usada como eixo y do gráfico.

A função retorna um objeto do matplotlib chamado `matplotlib.axes.Axes`, cuja documentação pode ser vista [aqui](https://matplotlib.org/stable/api/axes_api.html#matplotlib.axes.Axes).

Ilustremos a função com nossa tabela de dados criada.

```python
scatterplot(data = df, x = 'ValoresX', y = 'y1')
```

## **Parâmetros de Customização**

Antes de falarmos sobre outros parâmetros, vamos usar uma outra tabela que possui mais dados, que já vem embutida no seaborn. Para isso, basta chamar a função `load_dataset`.

```python
from seaborn import load_dataset
```

Em particular, usaremos o conjunto de dados "tips".

```python
tabela = load_dataset('tips')

tabela.sample(5)
```

Dentre os parâmetros de customização, vamos destacar os seguintes:

* `hue`: nome de uma coluna que servirá para distinguir pontos com diferentes cores. Por exemplo, se eu fizer o gráfico da conta total pela gorjeta, e querer diferenciar as gorjetas dadas para homens e mulheres, então colocaria `hue = 'sex'` como parâmetro.

* `style`: Ele possui praticamente a mesma funcionalidade do `hue`, só que ao invés de diferenciarmos por cores, ele diferencia pelo tipo de marcador, como ponto, cruz, x, etc.

Se combinarmos os dois parâmetros, ele diferenciará as categorias de maneira independente.

Vamos começar fazendo o gráfico de dispersão onde o eixo y será a gorjeta e o x será o valor total da conta. Aqui vamos distinguir os pontos pelo gênero do garçon.

```python
scatterplot(data = tabela, x = 'total_bill', y = 'tip', hue = 'sex')
```

Para separar entre homens que fumam ou não, e mulheres que fumam ou não, incluimos o parâmetro `style = 'smoker'`:

```python
scatterplot(data = tabela, x = 'total_bill', y = 'tip', hue = 'sex', style = 'smoker')
```
