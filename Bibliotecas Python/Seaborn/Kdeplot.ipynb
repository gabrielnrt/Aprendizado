# **`kdeplot()`**

## **Introdução**

Neste notebook falaremos sobre a função `kdeplot()` do seaborn, que é importada através do seguinte comando.

```python
from seaborn import kdeplot
```

Obs: sua documentação se encontra [neste link](https://seaborn.pydata.org/generated/seaborn.kdeplot.html). 

KDE significa **estimativa de densidade kernel**, que é uma estimativa da função densidade de probabilidade de uma variável aleatória.

## **Parâmetros Fundamentais**

Os parâmetros fundamentais dessa função são:

* `data`: array ou DataFrame que será referenciado para a realização das KDE's. Note que se inserirmos apenas este parâmetro, então o interpretador vai entender que é para fazer estimativas de densidade de probabilidade de todas as colunas da tabela.

* `x`: Vetor ou nome da coluna cujos dados estarão no eixo x.

* `y`: Vetor ou nome da coluna cujos dados estarão no eixo y. Este parâmetro é uma alternativa ao `x`.

Vale ressaltar que, se estamos calculando estimativas de densidade de probabilidade, então estamos trabalhando exclusivamente com atributos categóricos.

Para ilustrar, usaremos os dados de gorjetas, que estão na função `load_dataset`.

```python
from seaborn import load_dataset

tips = load_dataset('tips')
 
tips.dtypes
```

Usando a coluna `total_bill`, temos

```python
kdeplot(data = tips, x = 'total_bill')
```

## **Parâmetros Adicionais**

Note que podemos dividir a coluna da conta total em "conta total paga na horário do lanche", e "conta total paga no horário da janta", se considerarmos a coluna `time`. 

```python
tips.sample(5)
```

Assim, se quisermos fazer um KDE para a conta total paga em cada um desses períodos, então devemos inserir o seguinte parâmetro:

* `hue`: vetor ou nome da coluna que será usada para fazer distinção da coluna original.

```python
kdeplot(data = tips, x = 'total_bill', hue = 'time')
```
