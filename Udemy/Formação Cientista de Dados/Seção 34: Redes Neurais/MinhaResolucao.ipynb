# Resolução do problema "soybean.csv"

Aqui será uma espécie de rascunho da resolução do problema com a explicação. O código em si estará num arquivo com extensão ```.py```.

````python
from pandas import read_csv

df = read_csv('soybean.csv')

df
````

Como podemos ver, o arquivo possui 36 colunas, sendo que a classe é a última coluna, e as demais são características das plantas. Vamos primeiramente separar entre atributos e classe.

````python
x = df.iloc[:,:35].values

y = df['class'].values
````

Com isso, vamos usar o ```LabelEncoder()``` para transformar as colunas de string em números.

````python
from sklearn.preprocessing import LabelEncoder

# instancia0 = LabelEncoder()
# x[:,34]

for i in range(35):
    codificador = LabelEncoder()
    x[:,i] = codificador.fit_transform(x[:,i])

instancia = LabelEncoder()
y = instancia.fit_transform(y)
````

Agora vamos particionar os dados em treinamento e teste.

````python
from sklearn.model_selection import train_test_split

XTreino, XTeste, YTreino, YTeste = train_test_split(x,y,
                                                   test_size = 0.3,
                                                   random_state = 0)
````

Feito o tratamento dos dados, vamos implementar o modelo de Naive-Bayes propriamente dito.

````python
from sklearn.naive_bayes import GaussianNB

modelo = GaussianNB()

modelo.fit(XTreino,YTreino)

previsao = modelo.predict(XTeste)
````

Deste modo, vamos comparar os valores da previsao com os dados "de verdade".

````python
from sklearn.metrics import confusion_matrix, accuracy_score

confusao = confusion_matrix(previsao,YTeste)

confusao
````

````python
print(accuracy_score(previsao,YTeste))
````
