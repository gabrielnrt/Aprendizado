```python
from pandas import read_csv
from pylab import scatter, plot, title, xlabel, ylabel
from seaborn import regplot
```

```python
# Carregamento da base de dados
base = read_csv('/content/drive/MyDrive/Coisas em python/Udemy/Analise de Dados/Seção 13/dados/trees.csv')
base.head()
```

```python
# Gráfico de dispersão considerando o volume e a dispersão
# girth = circunferência

scatter(base.Girth, base.Volume, color = 'blue', facecolors = 'none', marker = 'o')

title('Árvores')
ylabel('Volume')
xlabel('Circunferência')
```

```python
# Gráfico de linha considerando o volume e o atributo "girth"
plot(base.Girth, base.Volume)
title('Árvores')
ylabel('Volume')
xlabel('Circunferência')
```

```python
# Gráfico de dispersão com 'afastamento' dos dados (jitter)
#fit_reg linha de tendência
regplot(base.Girth, base.Volume, data = base, x_jitter = 0.3, fit_reg = False)
```
