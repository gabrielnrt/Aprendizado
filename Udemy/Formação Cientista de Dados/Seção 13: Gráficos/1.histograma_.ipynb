## Fazendo o histograma

Num histograma pegamos uma amostra de dados e as dividimos em intervalos (bins), e assim colocamos no eixo y a frequência com a qual esse intervalo se repete.

Deste modo, para fazermos um histograma, os parâmetros essenciais a serem fornecidos são o conjunto de dados e quantas divisões queremos realizar. 

O comando para fazer o histograma é ```hist()```, da biblioteca matplotlib.pyplot

````python
hist(base.iloc[:,1], bins = 6)
title('Árvores')
ylabel('Frequência')
xlabel('Altura')
````

Podemos colocar outros parâmetros como cor por exemplo.

Mas como sabemos os valores que delimitam os intervalos do histograma? Para resolver esse problema, basta chamar a função ```histogram()``` da biblioteca numpy, colocando no argumento os dados a serem usados e o número de divisões.

````python
h = histogram(base.iloc[:,1], bins = 6)
h
````

Com isso, o primeiro array nos mostra a altura de cada barra, enquanto que o segundo array nos mostra os pontos que delimitam os intervalos de cada barra. 

No nosso caso, o primeiro intervalo vai de 63 a 67, o segundo vai de 67 a 71, e assim por diante.
