## Usando o distplot do seaborn

No seaborn, fazemos tanto um histograma quanto a curva da densidade com o mesmo comando, o ```distplot()```, e para especificar se queremos que ele retorne um histograma ou uma curva de densidade, usamos no argumento os parâmetros ```hist``` para histograma e ```kde``` -(gaussian) Kernel Density Estimate- para curva de densidade, ambos parâmetros boolenaos.


### Histograma

Para realizar um histograma, fazemos

````python
distplot(base['Height'], bins = 6, 
         hist = True, 
         kde = False)
````

Contudo, o gráfico está muito simplório, isto é, não temos uma linha para dividir cada gráfico e a cor padrão é meio "apagada". 

Para melhorar a aparência do nosso gráfico, introduzimos o parâmetro ```hist_kws```, que é um dicionário, e lá colocamos "color" para a cor das barras e "edgecolor" para a cor das linhas que separam cada barra. 

````python
distplot(base['Height'], bins = 6, 
         hist = True, hist_kws = {'color':'blue', 'edgecolor':'black'}, 
         kde = False)
````

### Curva de densidade

Se quisermos apenas uma curva de densidade, não precisamos informar o número de divisões pois o retorno é só a curva. Assim, para graficar a curva de densidade fazemos

````python
distplot(base['Height'],
         hist = False, 
         kde = True, color = 'navy')
````

Vale ressaltar que poderíamos colocar o parâmetro ```kde_kws```, análogo ao ```hist_kws```, mas como nesse caso estamos interessado mais em configurar a cor, basta colocarmos o parâmetro 'color':

````python
distplot(base['Height'],
         hist = False, 
         kde = True, kde_kws = {'color':'navy'})
````

### Densidade e Histograma juntos

Para termos os dois gráficos numa figura, basta configurar como 'True' nos parâmetros hist e kde.

````python
distplot(base['Height'], bins = 6,
         hist = True, hist_kws = {'color':'blue','edgecolor': 'black'},
         kde = True, color='orange')
````
